### 1. Loading five 128x128 pixel images, and printing out their RGB color channels as numpy arrays:

I chose 5 images depicting some animals. I'm simply showing what they look like:

```python
image_names = ['dog', 'cat', 'eagle', 'crocodile', 'axolotl']
images = [Image.open(path + '/' + i + '.png') for i in image_names]
for i in images:
    display(i)
```

And here I'm using the Python Image Library and Numpy to simply convert each of them into arrays containing the RGB value of each pixel; then I'm printing out the results:

```python
print('Array of pixels:\n')
for i in range(len(images)):
    p = images[i]
    pixels = np.array(p.getdata()).reshape(p.size[0], p.size[1], 3)
    print('IMAGE ' + str(i+1) + ':\n' + str(pixels) + '\n')
```

### 2. Calculating the average and the standard deviation of each channel, and normalizing with it. Printing out the exact values of average and standard deviations.

I used a double-for loop, for each image, and for each channel. First I showed each color channel separately, then I printed out the wanted values.

```python
channel_names = ['red', 'green', 'blue']
for i in range(len(images)):
    
    print('=================================')
    print('IMAGE ' + str(i+1) + ' color channels displayed:')
    print('=================================')
    p = images[i]
    pixels = np.array(p.getdata()).reshape(p.size[0], p.size[1], 3)
    
    for c in range(3):
    
        channel = np.array([[pixels[i][j][c] for j in range(p.size[1])] for i in range(p.size[0])])
        image_array = np.array([[[0 for e in range(c)] + [pixels[i][j][c]] + [0 for e in range(2-c)] 
                                 for j in range(p.size[1])] for i in range(p.size[0])])
        image = Image.fromarray(image_array.astype('uint8'), 'RGB')
        display(image)

        channel_avg = np.mean(channel)
        channel_var = np.std(channel)
        channel_normalized = (channel - channel_avg) / channel_var
        channel_normalized_avg = np.mean(channel_normalized)
        channel_normalized_var = np.std(channel_normalized)

        print('IMAGE ' + str(i+1) + ' '+ channel_names[c] + 
              ' channel average = ' + str(round(channel_avg,4)))
        print('IMAGE ' + str(i+1) + ' '+ channel_names[c] + 
              ' channel standard deviation (square root of the variance) = ' + str(round(channel_var,4)))

        print('NORMALIZED IMAGE ' + str(i+1) + ' '+ channel_names[c] + 
              ' channel average = ' + str(round(channel_normalized_avg,4)))
        print('NORMALIZED IMAGE ' + str(i+1) + ' '+ channel_names[c] + 
              ' channel standard deviation (square root of the variance) = ' 
              + str(round(channel_normalized_var,4)))

    print('\n')
```

### 3. Showing the text source of <a href='https://blog.keras.io/the-future-of-deep-learning.html'>"The future of deep learning" on the Keras Blog</a>:

I wrote a function that lets me gain access to the source code of a website:

```python
def source(url):
    
    hdr = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/69.0.3497.100 Safari/537.36',
       'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8',
       'Accept-Charset': 'ISO-8859-1,utf-8;q=0.7,*;q=0.3',
       'Accept-Encoding': 'none',
       'Accept-Language': 'en-US,en;q=0.8',
       'Connection': 'keep-alive'}
    
    request = urllib.request.Request(url,None,hdr)
    response = urllib.request.urlopen(request)
    data = response.read().decode('utf8')
    return data
website_source = source('https://blog.keras.io/the-future-of-deep-learning.html')
```

Then I've filtered all parts of the source code that were inside smaller/greater brackets (<, >). So only text inbetween shows up. And that worked perfectly:

```python
out = ''
inside_brackets = False
for i in website_source[website_source.find('<article>'):website_source.find('</article>')]:
    if(i == '<'):
        inside_brackets = True
    if(not inside_brackets):
        out += i
    if(i == '>'):
        inside_brackets = False
print(out)
```

Then I have deleted the indentations and long spaces, and lowered all characters. From the description of the exercise, it wasn't clear whether or not upper and lower characters of the same type counted separately or not; so here I count them as the same. (But obviously the `.lower()` part can be removed to count them separately.)

```python
out = out.replace('\n','').replace('  ','').lower()
out
```

Then I counted the occurence of each character, ordered them by frequency, and showed them in descending order on a graph. The most common characers were space, then `e`, `t` and `a`. The least common were digits and the `@` sign.

```python
character_dict = {}
for i in out:
    if i in character_dict:
        character_dict[i] += 1
    else:
        character_dict[i] = 1
        
character_frequencies = {char: num for char, num in
                         sorted(character_dict.items(), key=lambda item: item[1], reverse=True)}

plt.figure(figsize=(20,10))
plt.plot(list(character_frequencies.keys()),list(character_frequencies.values()))
plt.show()
```

### 4. Loading a sound file and showing its spectrogram

I loaded the first ~15 seconds on Vivaldi's Four Seasons and used Scipy's Wavefile data type to draw its spectrogram. (Though the natural logarithm of the spectrogram is way more visible, so I shwoed that. - This makes sense, because our perception is logarithmic.)

```python
rate, notes = wavfile.read('./audio/vivaldi.wav')
frequencies, times, spectrogram = signal.spectrogram(np.array([i[0] for i in notes], dtype=int), rate)

plt.figure(figsize=(20,10))
plt.pcolormesh(times, frequencies, np.log(spectrogram))
plt.ylabel('Frequency (Hz)')
plt.xlabel('Time (sec)')
plt.show()
```
